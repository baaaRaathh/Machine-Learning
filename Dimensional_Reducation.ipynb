```python
from tensorflow.keras.datasets import mnist
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from umap import UMAP
from umap.umap_ import UMAP
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test,y_test) = mnist.load_data()
print(x_train.shape, y_train.shape,x_test.shape,y_test.shape)
```

```python
x_train = x_train.reshape(len(x_train),-1)
x_train.shape
```

```python
xumap = UMAP(n_neighbors=10,min_dist=0.1,metric='correlation' )
um=xumap.fit(x_train)
```

```python
x_umap = UMAP(n_components=3,n_neighbors=10, min_dist=0.1, metric='correlation').fit_transform(x_train)
```

```python
plt.figure(figsize=(10, 8))
scatter = plt.scatter(x_umap[:, 0], x_umap[:, 1], c=y_train, cmap='Spectral', s=5, alpha=0.6)
plt.title('2D UMAP Projection of MNIST Dataset', fontsize=20)
plt.xlabel('UMAP Dimension 1', fontsize=14)
plt.ylabel('UMAP Dimension 2', fontsize=14)
cbar = plt.colorbar(scatter)
cbar.set_label('Digit Label', fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(10,8))
sc = plt.scatter(x_umap[:, 0], x_umap[:, 1], c=y_train)
plt.legend(handles=sc.legend_elements()[0], labels=list(range(10)))
plt.show()
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x_umap[:, 0], x_umap[:, 1], x_umap[:, 2],c=y_train, cmap='Spectral', s=2, alpha=0.6)
ax.set_title('3D UMAP Projection of MNIST Dataset', fontsize=16)
ax.set_xlabel('UMAP Dimension 1')
ax.set_ylabel('UMAP Dimension 2')
ax.set_zlabel('UMAP Dimension 3')
cbar = fig.colorbar(scatter, ax=ax, pad=0.1)
cbar.set_label('Digit Label', fontsize=12)
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming tsne is the result of t-SNE and y_train is your target labels

# Apply PCA to reduce tsne to 3D
pca = PCA(n_components=3)
tsne_3d = pca.fit_transform(tsne)  # Now tsne_3d has 3 components

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create a 3D scatter plot
sc = ax.scatter(tsne_3d[:, 0], tsne_3d[:, 1], tsne_3d[:, 2], c=y_train)

# Add a legend
ax.legend(handles=sc.legend_elements()[0], labels=list(range(10)))

plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

tsne_3d = np.column_stack((tsne, np.zeros(tsne.shape[0])))
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(tsne_3d[:, 0], tsne_3d[:, 1], tsne_3d[:, 2], c=y_train)
ax.legend(handles=sc.legend_elements()[0], labels=list(range(10)))

plt.show()
```

#Binaryclass Logistic Regresiion


```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#load the DataSet in the from the Scikitlearn

iris = load_iris()
fe = iris.data[:100,:] # we only take the frist 100 data in iris it give only 50-setosa,50-versinika
ta = iris.target[:100]
```

```python
#create  the scaling techinque
scale = StandardScaler()
fe_sc = scale.fit_transform(fe)
```

```python
#create the logidtic regressiom
log = LogisticRegression(random_state=0)
model = log.fit(fe_sc,ta)
```

##Formula


```python
#1/1+e^(w^t+B)
```

```python
test_data = [[.000000001233,.00000000351,.0000000134,.000000531]]
```

```python
model.predict(test_data)
```

```python
model.predict_proba(test_data)
```

```python
model.predict_log_proba(test_data)
```

#Reducing the varince using Regulariation


```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
feature = iris.data
target =iris.target

```

```python
scale = StandardScaler()
fe_sc = scale.fit_transform(feature)
```

```python
log_reg = LogisticRegressionCV(penalty='l2',Cs=10,random_state=0,n_jobs=-1)
```

```python
model=log_reg.fit(fe_sc,target)
```

```python
scale = StandardScaler()
fe_sc = scale.fit_transform(fe)
log = LogisticRegression(random_state=0,solver='sag')
model = log.fit(fe_sc,ta)
```
